### Q1. 마케팅 응답 고객들의 나이를 10살 단위로 변환 했을 때, 가장 많은 인원을 가진 나이대는?  (0\~9 : 0 , 10~19 : 10)

```python
age_n = (df.age //10).value_counts().index[0]
age_n*10
```

### Q3. 나이가 25살 이상 29살 미만인 응답 고객들중 housing컬럼의 값이 yes인 고객의 수는?

```python
age_25 = df.loc[(df.age>=25) & (df.age<29)]
len(age_25[age_25['housing']=='yes'])
```

### Q4. numeric한 값을 가지지 않은 컬럼들중 unique한 값을 가장 많이 가지는 컬럼은?

```python
df.info()
```

```python
nonu = df.iloc[:,2:]
nonu = nonu.drop(['balance','day','campaign','pdays','previous'], axis=1)
nonu
```

```python
result = nonu.nunique()
result = result.sort_values(ascending=False).index[0]
print(result)
```

### Q5. balance 컬럼값들의 평균값 이상을 가지는 데이터를 ID값을 기준으로 내림차순 정렬했을때 상위 100개 데이터의 balance값의 평균은?

```python
bal_mean = df.balance.mean()
```

```python
df[df['balance']>bal_mean].sort_values('ID', ascending=False).head(100).balance.mean()
```

```python
result = df[['age','balance']].corr().iloc[0,1]
print(result)
```

```python
marr = df.loc[(df['marital']=='divorced') | (df['marital']=='married')]
marr.groupby(['job','marital']).count()
```

```python
t = df.groupby(['job','marital']).size().reset_index()
pivotdf = t.pivot_table(index='job',columns='marital')[0]
pivotdf = pivotdf.fillna(0)
pivotdf
```

```python
pivotdf['ratio'] = pivotdf['divorced']/pivotdf['married']
pivotdf['ratio'].sort_values(ascending=False)[0]
```
